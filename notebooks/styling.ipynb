# Seaborn Styling Guide
Show how style, context, and axis options affect chart presentation.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
crash_df = sns.load_dataset("car_crashes")
crash_df.head()
```

```python
# Styling the plot
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# Plotting the distribution of 'not_distracted' variable
sns.displot(crash_df["not_distracted"], kde=True, )
```

```python
# Styling the plot
sns.set_style("darkgrid")
plt.figure(figsize=(10, 6))

# Plotting the distribution of 'not_distracted' variable
sns.displot(crash_df["not_distracted"], kde=True, )
```

```python
# Styling the plot
sns.set_style("white")
plt.figure(figsize=(10, 6))

# Plotting the distribution of 'not_distracted' variable
sns.displot(crash_df["not_distracted"], kde=True, )
```

```python
# Styling the plot
sns.set_style("dark")
plt.figure(figsize=(10, 6))

# Plotting the distribution of 'not_distracted' variable
sns.displot(crash_df["not_distracted"], kde=True, )
```

```python
# Styling the plot
sns.set_style("ticks")
plt.figure(figsize=(10, 6))

# Plotting the distribution of 'not_distracted' variable
sns.displot(crash_df["not_distracted"], kde=True, )
```

## Adjust Context and Labels

```python
# Styling the plot
sns.set_style("ticks")
plt.figure(figsize=(10, 6))

# SET the context to "notebook" with a larger font scale
#SET NOTEBOOK IS FOR PRESENTATIONS

sns.set_context("notebook", font_scale=1.5)

# Plotting the distribution of 'not_distracted' variable
sns.displot(crash_df["not_distracted"], kde=True, )
```

## Turn Off Chart Spines

### Without `sns.despine()`

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 4))
sns.set_context("paper", font_scale=1.5)

sns.jointplot(x="speeding", y="alcohol", data=crash_df, kind="reg")
```

### With `sns.despine()`

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 4))
sns.set_context("paper", font_scale=1.5)

sns.jointplot(x="speeding", y="alcohol", data=crash_df, kind="reg")

# Turn off the feature
sns.despine(left=True, bottom=True)
```

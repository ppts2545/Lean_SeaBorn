# Heatmap Analysis
Visualize correlation patterns from the `car_crashes` dataset with a readable matrix plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")
```

```python
crash_df = sns.load_dataset("car_crashes")
crash_df.head()
```

```python
sns.set_context("paper", font_scale=1.2)

crash_mx = crash_df.corr(numeric_only=True)

plt.figure(figsize=(9, 7))
sns.heatmap(crash_mx, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")

plt.title("Correlation Heatmap: Car Crashes Dataset", fontsize=13, pad=10)
plt.tight_layout()
plt.show()
```

# Pivot Table and Heatmap Analysis
This notebook demonstrates how to organize time-series data with a pivot table and visualize it with a heatmap.

## Workflow
1. Setup and import libraries
2. Load and preview dataset
3. Build pivot table
4. Visualize with heatmap
5. Interpret key patterns

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Optional: set a clean plotting style for consistency
sns.set_theme(style="white")
```

## Data Loading and Preview
Load the built-in `flights` dataset and quickly inspect its structure.

```python
flights_df = sns.load_dataset("flights")
flights_df.head()
```

## Pivot Table Construction
Reshape the dataset so rows are months, columns are years, and values are passenger counts.

## Pivot Table Result
Create and review the pivot table before visualization.

```python
flights_pivot = flights_df.pivot_table(
    index="month",
    columns="year",
    values="passengers"
)

flights_pivot
```

## Heatmap Visualization
Visualize the pivot table to identify monthly seasonality and year-over-year growth patterns.

```python
plt.figure(figsize=(12, 6))

sns.heatmap(
    flights_pivot,
    cmap="YlGnBu",
    linewidths=0.5,
    linecolor="white",
    annot=True,
    fmt=".0f"
)

plt.title("Monthly Airline Passengers by Year", fontsize=14, pad=12)
plt.xlabel("Year")
plt.ylabel("Month")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../Data/raws/ComputerSales.csv")
```

```python
tips_df = sns.load_dataset("tips")
```

```python
plt.figure(figsize=(8, 5))
sns.set_style("dark")
sns.set_context("talk")
sns.stripplot(x="day", y="total_bill", data=tips_df, hue="sex")
```

Paletes = you can change color make it more beautiful

you can search on matplotlib / colormaps docs and find theme of color you like

```python
plt.figure(figsize=(8, 5))
sns.set_style("dark")
sns.set_context("talk")
sns.stripplot(x="day", y="total_bill", data=tips_df, hue="sex", palette="magma")
```

```python
plt.figure(figsize=(8, 5))
sns.set_style("white")
sns.set_context("talk")
sns.stripplot(x="day", y="total_bill", data=tips_df, hue="sex", palette="afmhot", dodge=True)

plt.legend(loc=0)
```

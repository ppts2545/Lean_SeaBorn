```python
# Regression Plots (lmplot) Analysis

## Overview
**lmplot()** creates publication-ready regression plots with automatic confidence intervals and trend lines. It combines scatter plot + fitted regression line + optional faceting.

## When to Use lmplot
- ✅ Show linear relationships between two variables
- ✅ Add trend lines with confidence intervals (95% by default)
- ✅ Compare regressions across groups using `hue` or `col/row`
- ✅ Quick exploratory analysis of correlations
- ⚠️ For non-linear relationships, consider `regplot()` with polynomial fits

## Key Parameters
- `x, y`: Variables to plot
- `hue`: Color-code by categorical variable
- `col/row`: Create separate plots by category
- `scatter_kws`: Customize scatter points (size, alpha, etc.)
- `line_kws`: Customize regression line (width, color, etc.)
- `height/aspect`: Control plot size
```

```python
# Import Required Libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set theme for consistent styling
sns.set_theme(style="whitegrid", font_scale=1.0)
plt.rcParams['figure.dpi'] = 100
```

```python
# Load tips dataset
tips_df = sns.load_dataset("tips")
print(f"Shape: {tips_df.shape}")
tips_df.head()
```

```python
# Create regression plot with hue distinction by sex
fig = sns.lmplot(
    data=tips_df,
    x='total_bill', 
    y='tip', 
    hue='sex',
    palette=['steelblue', 'coral'],          # Male: steelblue, Female: coral
    scatter_kws={
        's': 100,                             # Point size
        'linewidths': 0.5, 
        'edgecolors': 'white', 
        'alpha': 0.7                          # Transparency
    },
    line_kws={'linewidth': 2.5},              # Regression line width
    height=6, 
    aspect=1.33
)

# Customize the plot
fig.set_axis_labels("Total Bill ($)", "Tip ($)", fontsize=12)
plt.suptitle('Tipping Behavior: Relationship between Bill & Tip by Gender', 
             fontsize=13, y=1.00, fontweight='bold')

# Add grid for better readability
fig.ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

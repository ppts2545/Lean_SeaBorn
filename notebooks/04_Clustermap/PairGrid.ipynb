```python
# Import Required Libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Set professional plotting style
sns.set_theme(style="whitegrid", font_scale=0.9)
plt.rcParams['figure.dpi'] = 100
sns.set_palette("husl")
```

```python
# Load iris dataset
iris = sns.load_dataset("iris")
iris.head()
```

```python
# Create PairGrid with scatter plots for all cells
iris_g = sns.PairGrid(iris, hue="species", height=2.5, aspect=1.0)
iris_g.map(plt.scatter, s=80, alpha=0.7, edgecolor='white', linewidth=0.5)
iris_g.add_legend(title="Iris Species", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.suptitle('Iris Dataset: Pairwise Relationships', y=1.00, fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Map different plots to diagonal vs off-diagonal
iris_g = sns.PairGrid(iris, hue="species", height=2.5, aspect=1.0)
iris_g.map_diag(plt.hist, bins=15, alpha=0.6, edgecolor='white', linewidth=0.5)
iris_g.map_offdiag(plt.scatter, s=80, alpha=0.7, edgecolor='white', linewidth=0.5)
iris_g.add_legend(title="Iris Species", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.suptitle('Distribution & Relationships', y=1.00, fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Explicit mapping for better control
iris_g = sns.PairGrid(iris, hue="species", height=2.5, aspect=1.0)
iris_g.map_diag(plt.hist, bins=15, alpha=0.6, edgecolor='white', linewidth=0.5)
iris_g.map_offdiag(plt.scatter, s=80, alpha=0.7, edgecolor='white', linewidth=0.5)
iris_g.add_legend(title="Iris Species")
plt.tight_layout()
plt.show()
```

```python
# Asymmetric grid: different plot types in upper vs lower triangles
iris_g = sns.PairGrid(iris, hue="species", height=2.5, aspect=1.0)
iris_g.map_diag(plt.hist, bins=15, alpha=0.6, edgecolor='white', linewidth=0.5)
iris_g.map_upper(plt.scatter, s=80, alpha=0.7, edgecolor='white', linewidth=0.5)
iris_g.map_lower(sns.kdeplot, fill=True, alpha=0.4, cmap="Set2")
iris_g.add_legend(title="Iris Species")
plt.suptitle('Mixed Visualization: Scatter (upper) + KDE (lower)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Custom relationship matrix - only show specific pairs
iris_g = sns.PairGrid(
    iris, 
    hue="species",
    x_vars=["sepal_length", "sepal_width"],    # Columns
    y_vars=["petal_length", "petal_width"],    # Rows
    height=3, 
    aspect=1.1
)

iris_g.map(plt.scatter, s=80, alpha=0.7, edgecolor='white', linewidth=0.5)
iris_g.add_legend(title="Iris Species")
plt.suptitle('Custom Relationship Matrix: Sepals vs Petals', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

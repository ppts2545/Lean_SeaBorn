# Distribution and Joint Plot Analysis
This notebook introduces univariate and bivariate distribution visualizations using seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
available_datasets = sns.get_dataset_names()
available_datasets[:10]
```

```python
crash_df = sns.load_dataset("car_crashes")
crash_df.head()
```

## Distribution Plot

Use this chart to analyze a single variable's distribution.

```python
sns.displot(data=crash_df, x="not_distracted", kde=True)
```

```python
# Increase bin count for more detail in the histogram.
sns.displot(data=crash_df, x="not_distracted", kde=True, bins=20)
```

## Joint Plot

Use this chart to compare two variables (scatter by default).

### Joint Plot with Regression

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

### Joint Plot with KDE

```python
sns.jointplot(x="speeding", y="alcohol", data=crash_df, kind="kde")
```

### Joint Plot with Hex Bins

```python
sns.jointplot(x="speeding", y="alcohol", data=crash_df, kind="hex")
```

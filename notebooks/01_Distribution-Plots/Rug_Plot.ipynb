# Rug Plot Analysis
Display individual observations of `tip` along a single axis.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
plt.figure(figsize=(8, 2.5))

sns.rugplot(data=tips_df, x="tip", height=0.2, color="darkred")

plt.title("Rug Plot of Tip", fontsize=13, pad=10)
plt.xlabel("Tip")
plt.yticks([])
plt.tight_layout()
plt.show()
```

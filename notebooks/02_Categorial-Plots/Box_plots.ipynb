# Box Plot Analysis
Use box plots to compare the bill distribution across days and customer sex.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
plt.figure(figsize=(9, 5))

sns.boxplot(x="day", y="total_bill", hue="sex", data=tips_df, palette="Set3")

plt.title("Total Bill Distribution by Day and Sex", fontsize=13, pad=10)
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.legend(title="Sex", loc="upper left")
plt.tight_layout()
plt.show()
```

# Strip Plot Analysis
Use strip plots to inspect individual bill values across categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
plt.figure(figsize=(8, 5))

sns.stripplot(x="day", y="total_bill", data=tips_df, palette="Set2")

plt.title("Strip Plot of Total Bill by Day", fontsize=13, pad=10)
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.stripplot(
    x="day",
    y="total_bill",
    data=tips_df,
    jitter=True,
    hue="sex",
    dodge=True,
    palette="Set1"
)

plt.title("Strip Plot with Sex as Hue", fontsize=13, pad=10)
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.legend(title="Sex", loc="upper left")
plt.tight_layout()
plt.show()
```

# Bar Plot Analysis
Compare average and custom statistical summaries of `total_bill` across categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="whitegrid")
```

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
plt.figure(figsize=(7, 4))
sns.barplot(x="sex", y="total_bill", data=tips_df)
plt.title("Mean Total Bill by Sex")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.barplot(x="sex", y="total_bill", data=tips_df, estimator=np.median)
plt.title("Median Total Bill by Sex")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.barplot(x="sex", y="total_bill", data=tips_df, estimator=np.std)
plt.title("Standard Deviation of Total Bill by Sex")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.barplot(x="sex", y="total_bill", data=tips_df, estimator=np.var)
plt.title("Variance of Total Bill by Sex")
plt.tight_layout()
plt.show()
```

```python
# Covariance needs two variables, so this cell is replaced with an additional useful summary.
plt.figure(figsize=(7, 4))
sns.barplot(x="sex", y="tip", data=tips_df, estimator=np.mean)
plt.title("Mean Tip by Sex")
plt.tight_layout()
plt.show()
```

## Custom Estimator Example
Define a custom metric and apply it in `sns.barplot`.

```python
my_func = lambda x: np.mean(x) + 2 * np.std(x)
```

```python
plt.figure(figsize=(7, 4))
sns.barplot(x="sex", y="total_bill", data=tips_df, estimator=my_func)
plt.title("Custom Estimator: mean + 2*std")
plt.tight_layout()
plt.show()
```
